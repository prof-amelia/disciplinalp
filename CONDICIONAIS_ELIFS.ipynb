```python
nota = 6

if nota >= 7:
  print ("Aprovado")
elif nota >= 5:
  print ("Recuperação")
else:
  print("Reprovado")
```

```python
nome = input("Digite o nome do aluno")
nota_prova = float(input("Digite a nota da prova:"))
nota_trabalho = float(input("Digite a nota do trabalho"))
media = (nota_prova + nota_trabalho)/2

if media >= 9:
  situacao = "Aprovado com excelência"
elif media >= 7:
  situacao = "Aprovado"
elif media >=5:
  situacao = "Recuperação"
else:
  situacao = "Reprovado"

print ("Aluno:", nome)
print ("Média:", media)
print ("Situação:", situacao)
```

```python
nome = input("Digite o nome do cliente: ")
renda = float(input("Digite a renda mensal: "))

if renda >= 10000:
    classificacao = "Cliente Premium Plus"

elif renda >= 8000:
    classificacao = "Cliente Premium"

elif renda >= 6000:
    classificacao = "Cliente Ouro"

elif renda >= 4000:
    classificacao = "Cliente Prata"

elif renda >= 2500:
    classificacao = "Cliente Bronze"

elif renda >= 1500:
    classificacao = "Cliente Básico"

elif renda >= 800:
    classificacao = "Cliente Baixa Renda"

else:
    classificacao = "Cliente de Risco"

print("Cliente:", nome)
print("Renda:", renda)
print("Classificação:", classificacao)
```
